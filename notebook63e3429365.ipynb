```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import f1_score, recall_score, roc_auc_score, confusion_matrix
import os

# Set the paths to your dataset
train_dir = r"C:\Users\ASUS\OneDrive\Desktop\gender 1"
test_dir = r"C:\Users\ASUS\OneDrive\Desktop\gender 1"

# Data Preprocessing
# Augment the training data and rescale all images
train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

# Rescale the test data (no augmentation for test data)
test_datagen = ImageDataGenerator(rescale=1./255)

# Load the training data from directory (50+50 images)
train_set = train_datagen.flow_from_directory(
    train_dir,
    target_size=(64, 64),  # Resize images to 64x64
    batch_size=16,         # You can change this
    class_mode='binary'    # Binary classification
)

# Load the test data from directory
test_set = test_datagen.flow_from_directory(
    test_dir,
    target_size=(64, 64),
    batch_size=16,
    class_mode='binary'
)

# Initialize the CNN
model = Sequential()

# First Convolutional Layer
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(64, 64, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Second Convolutional Layer
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Third Convolutional Layer
model.add(Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten the layer
model.add(Flatten())

# Fully Connected Layer
model.add(Dense(units=128, activation='relu'))
model.add(Dropout(0.5))  # Dropout to prevent overfitting

# Output Layer (sigmoid for binary classification)
model.add(Dense(units=1, activation='sigmoid'))

# Compile the CNN
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Print model summary
model.summary()

# Train the CNN on the training set
history = model.fit(
    train_set,
    epochs=25,
    validation_data=test_set
)
y_pred_prob = model.predict(test_set)
y_pred = (y_pred_prob > 0.5).astype(int)

y_true = test_set.classes

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(test_set)
print(f"Test Accuracy: {test_acc}")

auc = roc_auc_score(y_true, y_pred_prob)
print(f"AUC: {auc}")

f1 = f1_score(y_true, y_pred)
print(f"F1 Score: {f1}")

recall = recall_score(y_true, y_pred)
print(f"Recall: {recall}")

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
print(f"Confusion Matrix:\n{cm}")
```
